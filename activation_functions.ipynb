## 1-Sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

def der_sigmoid(x):
  return sigmoid(x)*(1-sigmoid(x))
```

```python
x=np.linspace(-10,10,100)
y=sigmoid(x)
dy=der_sigmoid(x)
```

```python
plt.plot(x,y,x,dy)
plt.title('Sigmoid & Derivative')
plt.legend(['sigmoid','der_sigmoid'])
plt.grid()
plt.show()
```

## 2-Tanh

```python
def tanh(x):
  return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

def der_tanh(x):
  return 1 - tanh(x)*tanh(x)
```

```python
x=np.linspace(-10,10,100)
y=tanh(x)
dy=der_tanh(x)
```

```python
plt.plot(x,y,x,dy)
plt.title('Tanh & Derivative')
plt.legend(['tanh','der_tanh'])
plt.grid()
plt.show()
```

## 3-Relu

```python
def ReLU(x):
  data=[max(0,value) for value in x]
  return np.array(data,dtype=float)
def der_ReLU(x):
  data=[1 if value > 0 else 0 for value in x]
  return np.array(data,dtype=float)
```

```python
x=np.linspace(-10,10,100)
y=ReLU(x)
dy=der_ReLU(x)
plt.plot(x,y,x,dy)
plt.title('ReLU & Derivative')
plt.legend(['ReLU','der_ReLU'])
plt.grid()
plt.show()
```

## 3-Softmax

```python
def softmax(x):
  return np.exp(x) / np.sum(np.exp(x),axis=0)
```

```python
x=np.linspace(-10,10,100)
y=softmax(x)
plt.plot(x,y)
plt.title('softmax')
plt.grid()
plt.show()
```
